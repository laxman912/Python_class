```python
# ATM Withdrawal Validator

# Taking inputs from the user
balance = int(input("Enter your current account balance (NPR): "))
daily_withdrawn = int(input("Enter total amount withdrawn today (NPR): "))
amount = int(input("Enter withdrawal amount (NPR): "))

# Checking conditions
if amount % 500 != 0:
    print("Invalid amount. Must be a multiple of NPR 500.")

elif amount > balance:
    print("Insufficient balance.")

elif daily_withdrawn + amount > 50000:
    print("Daily withdrawal limit reached.")

else:
    balance -= amount
    print("Withdrawal successful.")
    print("Your current balance after withdrawal: NPR", balance)
```

```python
# Online Store Discount System

# Taking input from the user
purchase_amount = float(input("Enter total purchase amount (NPR): "))
loyalty_member = input("Are you a loyalty member? (yes/no): ").lower()

# Applying discount based on purchase amount
if purchase_amount < 1000:
    discount = 0

elif purchase_amount < 5000:
    discount = 0.05   # 5%

elif purchase_amount < 15000:
    discount = 0.10   # 10%

else:
    discount = 0.20   # 20%

# Calculate amount after main discount
discounted_amount = purchase_amount - (purchase_amount * discount)

# Apply extra 5% discount for loyalty members
if loyalty_member == "yes":
    discounted_amount = discounted_amount - (discounted_amount * 0.05)

# Display results
print("Original Amount: NPR", purchase_amount)
print("Final Payable Amount: NPR", discounted_amount)
```

```python
# Inventory Restock Alert System

inventory = [
    {"item": "Rice", "stock": 5, "threshold": 10},
    {"item": "Eggs", "stock": 24, "threshold": 12},
    {"item": "Milk", "stock": 3, "threshold": 6},
    {"item": "Bread", "stock": 8, "threshold": 5},
    {"item": "Chicken", "stock": 0, "threshold": 4},
    {"item": "Cooking Oil", "stock": 2, "threshold": 3},
]

# Counter for items needing restock
restock_count = 0

# Loop through inventory
for product in inventory:
    
    # Check if stock is below threshold
    if product["stock"] < product["threshold"]:
        print("Restock Alert:", product["item"])
        restock_count += 1

# Display total items needing restock
print("\nTotal items needing restock:", restock_count)
```

```python
# Password Strength Checker

passwords = ["hello", "Hello123", "H3ll0@World", "12345678", "MyP@ss!"]

special_characters = "!@#$%^&*"

# Loop through each password
for password in passwords:
    
    print("\nChecking Password:", password)

    # Conditions
    has_upper = False
    has_lower = False
    has_digit = False
    has_special = False

    # Check each character
    for char in password:
        if char.isupper():
            has_upper = True
        elif char.islower():
            has_lower = True
        elif char.isdigit():
            has_digit = True
        elif char in special_characters:
            has_special = True

    # Store missing criteria
    missing = []

    if len(password) < 8:
        missing.append("At least 8 characters long")

    if not has_upper:
        missing.append("At least one uppercase letter")

    if not has_lower:
        missing.append("At least one lowercase letter")

    if not has_digit:
        missing.append("At least one digit")

    if not has_special:
        missing.append("At least one special character (!@#$%^&*)")

    # Final result
    if len(missing) == 0:
        print("Strong Password")
    else:
        print("Weak Password")
        print("Missing Criteria:")
        for item in missing:
            print("-", item)
```

```python
# Taxi Fare Calculator

trips = [
    {"distance": 1.5, "hour": 14},
    {"distance": 5.0, "hour": 22},
    {"distance": 12.0, "hour": 3},
    {"distance": 8.5, "hour": 10},
    {"distance": 2.0, "hour": 23},
]

# Loop through each trip
for i, trip in enumerate(trips, start=1):

    distance = trip["distance"]
    hour = trip["hour"]

    # Base fare
    fare = 150

    # Calculate fare based on distance
    if distance <= 2:
        fare = 150

    elif distance <= 10:
        extra_km = distance - 2
        fare = 150 + (extra_km * 35)

    else:
        # Fare for first 10 km
        fare = 150 + (8 * 35)

        # Additional distance beyond 10 km
        extra_km = distance - 10
        fare += extra_km * 28

    # Apply night surcharge (10 PM – 5 AM)
    if hour >= 22 or hour < 5:
        fare += fare * 0.10

    # Display result
    print(f"Trip {i}")
    print(f"Distance: {distance} km")
    print(f"Travel Hour: {hour}:00")
    print(f"Total Fare: NPR {fare:.2f}")
    print("----------------------")
```

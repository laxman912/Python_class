```python
# Create a class 
class Student:
    # write its atributes
    def __init__(self, name, grade, student_id):
        self.name = name
        self.grade = grade
        self.student_id = student_id

        def display_info(self):
            print(f"Name: {self.name}")
            print(f"Grade: {self.grade}")
            print(f"Student ID: {self.student_id}")
            print("-" *30)
```

```python
# Expense Sharing Program

print("Welcome to the Expense Sharing Program")
```

```python
# Function to store group members

def get_members():
    members = []
    
    total_members = int(input("Enter number of group members: "))
    
    for i in range(total_members):
        name = input(f"Enter name of member {i+1}: ")
        members.append(name)
    
    return members
```

```python
# Function to store expenses paid by each member

def get_expenses(members):
    expenses = {}
    
    for member in members:
        amount = float(input(f"Enter expense paid by {member}: "))
        expenses[member] = amount
    
    return expenses
```

```python
# Function to calculate total expense

def calculate_total(expenses):
    total = sum(expenses.values())
    return total
```

```python
# Function to calculate equal share

def calculate_share(total, number_of_members):
    share = total / number_of_members
    return share
```

```python
# Function to show who paid extra and who needs to pay

def show_balance(expenses, share):
    
    print("\nExpense Summary")
    print("-------------------------")
    
    for member, amount_paid in expenses.items():
        
        difference = amount_paid - share
        
        if difference > 0:
            print(f"{member} paid extra NPR {difference:.2f}")
        
        elif difference < 0:
            print(f"{member} needs to pay NPR {abs(difference):.2f}")
        
        else:
            print(f"{member} paid exact amount")
```

```python
# Main program

members = get_members()

expenses = get_expenses(members)

total_expense = calculate_total(expenses)

equal_share = calculate_share(total_expense, len(members))

print("\nTotal Expense:", total_expense)
print("Equal Share Per Person:", round(equal_share, 2))

show_balance(expenses, equal_share)
```

```python
# Mini Banking App for Students

print("Welcome to the Mini Banking App")
```

```python
# Dictionary to store student accounts

accounts = {}
```

```python
# Function to create account

def create_account():
    
    name = input("Enter student name: ")
    
    if name in accounts:
        print("Account already exists!")
    
    else:
        deposit = float(input("Enter initial deposit: "))
        
        accounts[name] = deposit
        
        print("Account created successfully!")
```

```python
# Function to deposit money

def deposit_money():
    
    name = input("Enter account holder name: ")
    
    if name in accounts:
        
        amount = float(input("Enter deposit amount: "))
        
        accounts[name] += amount
        
        print("Money deposited successfully!")
    
    else:
        print("Account not found!")
```

```python
# Function to withdraw money

def withdraw_money():
    
    name = input("Enter account holder name: ")
    
    if name in accounts:
        
        amount = float(input("Enter withdrawal amount: "))
        
        minimum_balance = 500
        
        if accounts[name] - amount >= minimum_balance:
            
            accounts[name] -= amount
            
            print("Withdrawal successful!")
        
        else:
            print("Minimum balance of NPR 500 must be maintained!")
    
    else:
        print("Account not found!")
```

```python
# Function to withdraw money

def withdraw_money():
    
    name = input("Enter account holder name: ")
    
    if name in accounts:
        
        amount = float(input("Enter withdrawal amount: "))
        
        minimum_balance = 500
        
        if accounts[name] - amount >= minimum_balance:
            
            accounts[name] -= amount
            
            print("Withdrawal successful!")
        
        else:
            print("Minimum balance of NPR 500 must be maintained!")
    
    else:
        print("Account not found!")
```

```python
# Function to check balance

def check_balance():
    
    name = input("Enter account holder name: ")
    
    if name in accounts:
        
        print(f"Current balance of {name}: NPR {accounts[name]:.2f}")
    
    else:
        print("Account not found!")
```

```python
# Function to transfer money

def transfer_money():
    
    sender = input("Enter sender name: ")
    receiver = input("Enter receiver name: ")
    
    if sender in accounts and receiver in accounts:
        
        amount = float(input("Enter transfer amount: "))
        
        minimum_balance = 500
        
        if accounts[sender] - amount >= minimum_balance:
            
            accounts[sender] -= amount
            accounts[receiver] += amount
            
            print("Money transferred successfully!")
        
        else:
            print("Transfer failed! Minimum balance required.")
    
    else:
        print("One or both accounts not found!")
```

```python
# Function to display all account holders

def view_accounts():
    
    if len(accounts) == 0:
        print("No accounts available.")
    
    else:
        print("\nAccount Holders")
        print("-------------------")
        
        for name in accounts:
            print(name)
```

```python
# Main program with menu

while True:
    
    print("\n===== Banking Menu =====")
    print("1. Create Account")
    print("2. Deposit Money")
    print("3. Withdraw Money")
    print("4. Check Balance")
    print("5. Transfer Money")
    print("6. View All Account Holders")
    print("7. Exit")
    
    choice = input("Enter your choice: ")
    
    if choice == "1":
        create_account()
    
    elif choice == "2":
        deposit_money()
    
    elif choice == "3":
        withdraw_money()
    
    elif choice == "4":
        check_balance()
    
    elif choice == "5":
        transfer_money()
    
    elif choice == "6":
        view_accounts()
    
    elif choice == "7":
        print("Thank you for using Mini Banking App!")
        break
    
    else:
        print("Invalid choice! Please try again.")
```

```python
# Nepal Election Vote Counting System

print("Welcome to Nepal Election Vote Counting System")
```

```python
# Dictionary to store votes

votes = {
    "Nepali Congress": 0,
    "CPN-UML": 0,
    "Maoist Centre": 0,
    "Rastriya Swatantra Party": 0,
    "Rastriya Prajatantra Party": 0,
    "Janata Samajbadi Party": 0
}
```

```python
# Function to display parties

def display_parties():
    
    print("\n===== Political Parties =====")
    
    parties = list(votes.keys())
    
    for index, party in enumerate(parties, start=1):
        print(f"{index}. {party}")
```

```python
# Function to cast vote

def cast_vote():
    
    display_parties()
    
    parties = list(votes.keys())
    
    choice = int(input("Enter party number: "))
    
    if 1 <= choice <= len(parties):
        
        selected_party = parties[choice - 1]
        
        votes[selected_party] += 1
        
        print(f"Vote successfully given to {selected_party}")
    
    else:
        print("Invalid party number!")
```

```python
# Function to display vote count

def view_vote_count():
    
    print("\n===== Vote Count =====")
    
    for party, total_votes in votes.items():
        print(f"{party} : {total_votes} votes")
```

```python
# Function to show winner or equal votes

def show_winner():
    
    highest_votes = max(votes.values())
    
    winners = []
    
    for party, total_votes in votes.items():
        
        if total_votes == highest_votes:
            winners.append(party)
    
    print("\n===== Election Result =====")
    
    # No votes case
    if highest_votes == 0:
        print("No votes have been cast yet!")
    
    # Single winner case
    elif len(winners) == 1:
        print(f"Winner: {winners[0]}")
        print(f"Total Votes: {highest_votes}")
    
    # Equal votes case
    else:
        print("Tie between following parties:")
        
        for party in winners:
            print(party)
        
        print(f"Each party received {highest_votes} votes")
```

```python
# Main menu function

def main_menu():
    
    while True:
        
        print("\n========== Election Menu ==========")
        print("1. Cast Vote")
        print("2. View Vote Count")
        print("3. Show Winner")
        print("4. Exit")
        
        choice = input("Enter your choice: ")
        
        # Cast Vote
        if choice == "1":
            cast_vote()
        
        # View Vote Count
        elif choice == "2":
            view_vote_count()
        
        # Show Winner
        elif choice == "3":
            show_winner()
        
        # Exit
        elif choice == "4":
            print("Thank you for using the Voting System!")
            break
        
        else:
            print("Invalid choice! Please try again.")
```

```python
# Run the election system

main_menu()
```

```python
inventory = {
    "rice": {"price": 120, "stock": 20},
    "milk": {"price": 90, "stock": 10},
    "bread": {"price": 60, "stock": 15},
    "eggs": {"price": 15, "stock": 30}
}

cart = {
    "rice": 2,
    "milk": 3,
    "eggs": 12
}
```

```python
def check_stock(inventory, item, quantity):
    """
    Checks whether enough stock is available.
    Returns True if available, otherwise False.
    """
    return inventory[item]["stock"] >= quantity
```

```python
def calculate_item_cost(price, quantity):
    """
    Calculates cost of a single item.
    """
    return price * quantity
```

```python
def update_stock(inventory, item, quantity):
    """
    Reduces stock after purchase.
    """
    inventory[item]["stock"] -= quantity
```

```python
def print_bill(bill_items, grand_total):
    """
    Prints the final bill.
    """
    print("---- Bill ----")

    for item, quantity, cost in bill_items:
        print(f"{item} x{quantity} = NPR {cost}")

    print(f"Grand Total: NPR {grand_total}")
    print("--------------")
```

```python
def print_inventory(inventory):
    """
    Displays updated stock.
    """
    print("\nUpdated Stock:")

    for item, details in inventory.items():
        print(f"{item} = {details['stock']}")
```

```python
def process_order(inventory, cart):

    grand_total = 0
    bill_items = []

    for item, quantity in cart.items():

        if item not in inventory:
            print(f"{item} not found in inventory.")
            continue

        if check_stock(inventory, item, quantity):

            price = inventory[item]["price"]

            cost = calculate_item_cost(price, quantity)

            grand_total += cost

            bill_items.append((item, quantity, cost))

            update_stock(inventory, item, quantity)

        else:
            print(f"Sorry, not enough stock for {item}")

    print_bill(bill_items, grand_total)

    print_inventory(inventory)
```

```python
process_order(inventory, cart)
```

```python
# List containing sensor location and water level in metres
sensors = [
    ("Chatara", 2.8),
    ("Tribeni Ghat", 5.4),
    ("Koshi Barrage", 4.1),
    ("Sunsari Bridge", 1.9),
    ("Saptakoshi Camp", 6.0)
]
```

```python
# Function to determine flood risk based on water level
def check_water_level(location, level_metres):

    # Water level below 3 metres
    if level_metres < 3:
        return "Safe"

    # Water level between 3 and 5 metres
    elif 3 <= level_metres <= 5:
        return "Warning — Alert nearby villages"

    # Water level above 5 metres
    else:
        return "DANGER — Evacuate immediately!"
```

```python
# Loop through each sensor and display its status
for location, level in sensors:

    # Call the function and store the result
    status = check_water_level(location, level)

    # Print the location, water level, and status
    print(f"{location} ({level} m): {status}")
```

```python
# Nepali month names
bs_months = [
    "Baisakh", "Jestha", "Ashadh", "Shrawan",
    "Bhadra", "Ashwin", "Kartik", "Mangsir",
    "Poush", "Magh", "Falgun", "Chaitra"
]

# Customer records
customers = [
    {"name": "Ramesh Thapa", "date": "1985-06-24", "cal": "AD", "need": "BS", "style": "full"},
    {"name": "Sunita Karki", "date": "2055-09-10", "cal": "BS", "need": "AD", "style": "iso"},
    {"name": "Bikash Rai", "date": "1998-11-30", "cal": "AD", "need": "BS", "style": "nepali"},
    {"name": "Anjali Gurung", "date": "2040-01-05", "cal": "BS", "need": "AD", "style": "full"}
]
```

```python
# Function to convert date between AD and BS
def convert_date(date_str, from_cal, to_cal):

    # If both calendars are same, return original date
    if from_cal == to_cal:
        return date_str

    # Split date into year, month, and day
    year, month, day = map(int, date_str.split("-"))

    # AD to BS conversion
    if from_cal == "AD" and to_cal == "BS":
        year += 56

    # BS to AD conversion
    elif from_cal == "BS" and to_cal == "AD":
        year -= 56

    # Return converted date in YYYY-MM-DD format
    return f"{year:04d}-{month:02d}-{day:02d}"
```

```python
# Function to format converted date
def format_date(date_str, calendar, style):

    year, month, day = map(int, date_str.split("-"))

    # ISO style
    if style == "iso":
        return f"{date_str} {calendar}"

    # Full style
    elif style == "full":

        # Use Nepali month names for BS dates
        if calendar == "BS":
            month_name = bs_months[month - 1]
            return f"{day}th {month_name}, {year} BS"

        # Use normal date for AD
        return f"{day:02d}-{month:02d}-{year} AD"

    # Nepali style
    elif style == "nepali":

        if calendar == "BS":
            month_name = bs_months[month - 1]
            return f"{day} {month_name} {year} BS"

        return f"{date_str} AD"

    # Default
    return date_str
```

```python
# Process every customer
for customer in customers:

    # Convert date
    converted_date = convert_date(
        customer["date"],
        customer["cal"],
        customer["need"]
    )

    # Format converted date
    formatted_date = format_date(
        converted_date,
        customer["need"],
        customer["style"]
    )

    # Print result
    print(
        f"{customer['name']} | "
        f"Original: {customer['date']} {customer['cal']} | "
        f"Converted: {formatted_date}"
    )
```

```python
# News article text
text = """
Nepal is a beautiful country. Nepal has Mount Everest.
Everest is the highest mountain in the world.
Many tourists visit Nepal every year to see Everest
and other mountains. Nepal is known for its
mountains and natural beauty.
"""
```

```python
# Function to count word frequencies and return top 3 words
def word_frequency(text):

    # Remove punctuation marks
    text = text.replace(".", "")
    text = text.replace(",", "")

    # Convert all text to lowercase
    text = text.lower()

    # Split text into individual words
    words = text.split()

    # Dictionary to store word counts
    frequency = {}

    # Count occurrences of each word
    for word in words:

        if word in frequency:
            frequency[word] += 1
        else:
            frequency[word] = 1

    # Sort dictionary by frequency in descending order
    top_words = sorted(
        frequency.items(),
        key=lambda item: item[1],
        reverse=True
    )

    # Return top 3 most frequent words
    return top_words[:3]
```

```python
# Get top 3 words
result = word_frequency(text)

# Display result
print("Top 3 words:")

for word, count in result:
    print(f"{word} — {count} times")
```

```python
# Dictionary storing account information
accounts = {
    "A001": {"name": "Ramesh Thapa", "balance": 15000, "pin": "1234"},
    "A002": {"name": "Sunita Karki", "balance": 8500, "pin": "5678"},
    "A003": {"name": "Bikash Rai", "balance": 22000, "pin": "9012"}
}
```

```python
# Function to perform ATM operations
def atm(account_id, pin, action, amount=0):

    # Check whether account exists
    if account_id not in accounts:
        print("Account not found")
        return

    # Check whether PIN is correct
    if accounts[account_id]["pin"] != pin:
        print("Incorrect PIN")
        return

    # Get account details
    account = accounts[account_id]

    # Check balance
    if action == "balance":
        print(f"Account Holder: {account['name']}")
        print(f"Current Balance: NPR {account['balance']}")

    # Deposit money
    elif action == "deposit":
        account["balance"] += amount

        print(f"NPR {amount} deposited successfully.")
        print(f"New Balance: NPR {account['balance']}")

    # Withdraw money
    elif action == "withdraw":

        # Check sufficient balance
        if amount > account["balance"]:
            print("Insufficient funds")

        else:
            account["balance"] -= amount

            print(f"NPR {amount} withdrawn successfully.")
            print(f"Remaining Balance: NPR {account['balance']}")

    # Invalid action
    else:
        print("Invalid action")
```

```python
# Check balance
atm("A001", "1234", "balance")

print()

# Wrong PIN
atm("A002", "0000", "withdraw", 2000)

print()

# Deposit money
atm("A002", "5678", "deposit", 3000)

print()

# Insufficient funds
atm("A003", "9012", "withdraw", 25000)

print()

# Account not found
atm("A004", "1111", "balance")
```
